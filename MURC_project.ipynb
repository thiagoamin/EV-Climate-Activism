## Previous research

Lubell’s Collective Action Model
Attempting to create a theoretical model that predicts the expected value of participation
in climate activism, Lubell developed a collective interest model. His multivariable model
contains the following parameters: two probability variables which represent the group’s
probability of success ($p_{g}$) and an agent’s marginal contribution to this probability of success
($p_{i}$); selective costs ($C$) and benefits ($B$) to the individual; and the value of the collective public
good ($V$). The relationship between these variables is described in the following equation:

$EV(Global Warming Activism) =  p_{g} * p_{i} * V - C + B  $

Nevertheless, there are a few significant problems with the model described above. Firstly,
the author does not distinguish between success-dependent and success-independent benefits
and costs. For example, consider the two benefits: a heightened sense of collective social
identity and political power resulting from the group’s success. While the former will be yielded
regardless of the group’s success, the latter will only come to place if the group accomplishes its
goals. Therefore, resulting political power is dependent on, and its value should be affected by,
the perceived probability that the group will succeed.

Secondly, the value of the collective good (V) definition is ambiguous. The author fails to
elaborate on whether the variable represents the value at an individual or global level.
Nonetheless, this distinguishment would drastically impact the range of values obtained for this
variable. Looking at V at an individual level, the range obtained would be comparable to the
one obtained for selective costs and benefits. On the other hand, if V is seen as the
movement’s significance globally, the values it would take would be vastly higher than those
produced by the selective incentive variables.

Furthermore, an additional problem would arise from the interpretation of V as the
societal value of the collective good. Such reading would imply that agents do not act purely on
self-interest and that humans may perform actions of altruistic nature. The altruistic
supposition would conflict with the theory of rational choice, which states that agents always
make decisions that maximize personal utility. If individuals did act altruistically, it could be
concluded that individuals act irrationally under the theory of rational choice, defeating the
purpose of constructing a model to predict their rational behaviour.

Finally, the model also yields counterintuitive results in many scenarios. The variables
($p_{g} * p_{i}$) can be seen as a coefficient for V, adjusting its weight accordingly in cases like the
free-rider issue. V is a success-dependent variable (it is only yielded if the group succeeds), and
thus it is only logical that the probabilities will determine its weighting in the expected value
function. Since V can be interpreted in two different manners, I will examine different scenarios
under the two different interpretations.

First, let us consider that V is the value of the collective good for the individual. Consider
the scenario where the probability of group success ($P_{g}$) is 0.9, and the agent’s marginal
contribution to this probability of success ($P_{i}$) is 0.5. In other words, the individual will cause a
significant increase in the group’s probability of success, and without him/her, the group is
unlikely to succeed. In this situation, it would only be rational for the agent to consider the
value of $V$ close to its “unweighted” value (which assumes a 100% chance of success). However,
only 0.45 (45%) of the original $V$ value is accounted for, which underrepresents the importance
of $V$ in the expected value for this specific scenario. Graph 1 shows the relationship between
($p_{g} * p_{i}$) and $p_{i}$ for three different fixed values of $p_{g}$, exposing that $V$ is underweighted for
many combinations of $p_{g}$ and $p_{i}$



```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.linspace(0,1,100)
pg = [0.9,0.8,0.7]

for g in pg:
  y = pi * g
  plt.plot(pi,y,label= 'pg = ' + str(g))

plt.legend()
plt.ylim(0,1)
plt.xlabel('Additive contribution of individual (pi)')
plt.ylabel('pg * pi')
```

Finally, let us assume that $V$ is the value of the collective good for society (global scale). It is
only reasonable to assume that the variable would take on enormous values (thousands and
hundreds of thousands). In this case, $V$ would constantly be overweighted in the expected value
equation, as it is shown below. Graph 2 shows the relationship between $p_{g} * p_{i}$ and
$p_{g} * p_{i} * V$ for values of $V$ ranging from 1,000 to 10,000. Even for low values of $p_{g} * p_{i}$,
$p_{g} * p_{i} * V$ still yields high values, which makes the term have big impact on the resulting
expected value. This would lead to the free-rider problem, as the value of the collective good still
influences the expected value regardless of the probability variables.

```python
x = np.linspace(0,1,100)

for i in [1000,5000,10000]:
  y = i*x
  plt.plot(x,y,label='V = ' + str(i))

plt.xlim(0,1)
plt.legend()
```

## Linking measurements with theoretical variables

In each case, we need the appropriate function to convert the units in which a variable is measured to the units in the theoretical model. First, $P_{i}$ is a probability. Probabilities range from 0 to 1. But the variable is measured using a Likert scale, which ranges from 1 to 5. So we need a way to convert . Let's say $x$ the Likert score.

Our starting point is to use a linear regression.

$P_{i} = a + b*x$

We can convert any linear equation on the real number line into an equation that fits on the probability scale with the logistic function. It looks imitiating. But it's merely measuring what proportion of the maximium value has the linear regression travelled across. When the linear regression is big, the difference between the top and bottom of the fraction is small so the total value is close to one. When the linear regression is small, the difference will be big and the probability will be close to zero.

$P_{i} = \frac{exp(a + b*x)}{1 + exp(a + b*x)} $

```python
x = np.array([1,2,3,4,5])
xc = x - np.mean(x) # we have to center the x axis so Likert 3 corresponds to 0
a = 0
b = 2

y = np.exp(a + b*xc) / (1 + np.exp(a + b*xc))

plt.plot(x,y)
plt.xlabel('Likert scale answer')
plt.ylim(0,1)
plt.xticks([1,2,3,4,5])
plt.ylabel('P_i')
```

We also have variables that are measured by multiple questions. Those suggest we need a multi-variate regression design.

$P_{g} = a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3} $

Each of $x_{1},x_{2},x_{3}$ are measured by a Likert scale and a survey question. Adjusted to the probability scale, we end up with:

$P_{g} = \frac{exp(a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3})}{1 + exp(a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3})} $

The benefits and costs are measured on the real number line, which makes things a bit easier. One of the $B_{i}$ measures is a true/false question. This means we can treat $x_{2}$ as indicator variable where it takes on values 0 and 1.

$B_{i} = a + b_{1}*x_{1} + b_{2}*x_{2} $

Similar functions can be built for the remaining variables.

$B_{d} = a + b_{1}*x_{1}$

$ C_{i} = a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3}$

$ C_{d} = a + b_{1}*x_{1} + b_{2}*x_{2}$

## Complete regression model

$$ev(activism) = P_{i} * (1 - P_{g}) * (B_{d} - C_{d})) + B_{i} - C_{i}$$

replace every theoretical variable with the measure variable.

$$ev(activism) = \frac{exp(a + b*x)}{1 + exp(a + b*x)} * (1 - \frac{exp(a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3})}{1 + exp(a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3})}) * ((a + b_{1}*x_{1}) - (a + b_{1}*x_{1} + b_{2}*x_{2})) + (a + b_{1}*x_{1} + b_{2}*x_{2}) - (a + b_{1}*x_{1} + b_{2}*x_{2} + b_{3}*x_{3})$$

Replace the subscripts so all the terms are unique.

$$ev(activism) = \frac{exp(a_{1} + b_{1}*x_{1})}{1 + exp( a_{1} +b_{1}*x_{1})} * (1 - \frac{exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}{1 + exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}) * ((a_{3} + b_{5}*x_{5}) - (a_{4} + b_{6}*x_{6} + b_{7}*x_{7})) + (a_{5} + b_{8}*x_{8} + b_{9}*x_{9}) - (a_{6} + b_{10}*x_{10} + b_{11}*x_{11} + b_{12}*x_{12})$$

Drop parenthesis and redistribute the minus signs.

$$ev(activism) = \frac{exp(a_{1} + b_{1}*x_{1})}{1 + exp( a_{1} +b_{1}*x_{1})} * (1 - \frac{exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}{1 + exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}) * (a_{3} + b_{5}*x_{5} - a_{4} - b_{6}*x_{6} - b_{7}*x_{7}) + a_{5} + b_{8}*x_{8} + b_{9}*x_{9} - a_{6} - b_{10}*x_{10} - b_{11}*x_{11} - b_{12}*x_{12})$$

We can reduce the number of independent parameters by replacing $a_{1} - a_{2}$ type terms with a single constant. Regression techniques will work fine estimating two intercepts or one. But generally models with fewer parameters are better behaved for model fitting software.

$$ev(activism) = \frac{exp(a_{1} + b_{1}*x_{1})}{1 + exp( a_{1} +b_{1}*x_{1})} * (1 - \frac{exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}{1 + exp(a_{2} + b_{2}*x_{2} + b_{3}*x_{3} + b_{4}*x_{4})}) * (a_{3} + b_{5}*x_{5} - b_{6}*x_{6} - b_{7}*x_{7}) + a_{4} + b_{8}*x_{8} + b_{9}*x_{9} - b_{10}*x_{10} - b_{11}*x_{11} - b_{12}*x_{12})$$

### Better idea

Here we test for identifiability. We'll build simulated data sets using both model structures (ours and Lobell). Then we'll fit each model to each data set. We'll compute a Bayes factor. When the underlying simulation matches the Lubell model, it should have the higher Bayes factor. When the underlying simulation uses our structure, then the Lubell model should have a low Bayes factor.

```python
def simulate_person():
  pi = stats.uniform().rvs()
  pg = stats.uniform().rvs()
  bd = stats.norm(500,200).rvs()
  cd = stats.norm(100,30).rvs()
  bi = stats.norm(10,5).rvs()
  ci = stats.norm(50,20).rvs()

  ev = pi*pg*(bd-cd) + bi - ci

  return [ev,pi,pg,bd,cd,bi,ci]

t = [simulate_person() for i in range(500)]
t = np.array(t)

import pandas as pd

names = ['ev',
         'pi',
         'pg',
         'bd',
         'cd',
         'bi',
         'ci']

data = pd.DataFrame(t,columns=names)
data
```

```python
# model with no measurement problems with Lubell structure

y = data['ev']
x1 = data['pi']
x2 = data['pg']
x3 = data['bd']
x4 = data['cd']
x5 = data['bi']
x6 = data['ci']

with pm.Model() as m0_direct:

  # intercept priors

  α1 = pm.Normal('α1',0,0.5)
  α2 = pm.Normal('α2',0,0.5)
  α3 = pm.Normal('α3',0,0.5)
  α4 = pm.Normal('α4',0,0.5)

  # slope priors

  β1 = pm.Normal('β1',0,0.5)
  β2 = pm.Normal('β2',0,0.5)
  β3 = pm.Normal('β3',0,0.5)
  β4 = pm.Normal('β4',0,0.5)
  β5 = pm.Normal('β5',0,0.5)
  β6 = pm.Normal('β6',0,0.5)

  # error prior

  ϵ = pm.Exponential('ϵ',1)

  pi = pm.Deterministic('pi', np.exp(α1 + β1*x1) / (1 + np.exp(α1 + β1*x1)))
  pg = pm.Deterministic('pg', np.exp(α2 + β2*x2) / (1 + np.exp(α2 + β2*x2)))
  bd = pm.Deterministic('bd - cd', α3 + β3*x3 + β4*x4)
  bi = pm.Deterministic('bi - ci', α4 + β5*x5 + β6*x6)

  μ = pm.Deterministic('μ', pi*pg*bd + bi)

  y_pred = pm.Normal('ev', μ, ϵ, observed = y)

  trace_m0_direct = pm.sample(1000)
```
